RENAME OR REORDER COLUMNS

```python
df.rename(columns={"Customer ID":'CustomerID'},inplace=True)
df.rename(columns={"Transaction ID":'TransactionID'},inplace=True)
df.rename(columns={"Price per Unit":'Price'},inplace=True)
df.rename(columns={"Total Amount":'TotalAmount'},inplace=True)
```

```python
df=df[["TransactionID","CustomerID","Date","Gender","Age","Product Category","Quantity","Price","TotalAmount"]] #Re-Ordering the columns
```

```python
df.head()
```

CHANGE THE DATA TYPES

```python
df['Age']=df['Age'].astype(int)
df['CustomerId']=df['CustomerID'].astype(object)
df['Date']=pd.to_datetime(df['Date'])
df['Gender']=df['Gender'].astype(object)
df['Quantity']=df['Quantity'].astype(float)
df['Price']=df['Price'].astype(float)
df['TotalAount']=df['TotalAmount'].astype(float)
```

```python
print("Min Age:",df['Age'].min())
print("Max Age:",df['Age'].max())
print("Mean Age:",df['Age'].mean())
print("Median:",df['Age'].median())
print("Mode:",df['Age'].mode())
print("standard deviation:",df['Age'].std())
```

```python
print("Min Price:",df['Price'].min())
print("Max Price:",df['Price'].max())
print("Mean :",df['Price'].mean())
print("Median:",df['Price'].median())
print("Mode:",df['Price'].mode())
print("standard deviation:",df['Price'].std())
```

```python
print("Min Quantity:",df['Quantity'].min())
print("Max Quantity:",df['Quantity'].max())
print("Mean :",df['Quantity'].mean())
print("Median :",df['Quantity'].median())
print("Mode:",df['Age'].mode())
print("standard deviation:",df['Quantity'].std())
```

```python
print("Min TotalAmount:",df['TotalAmount'].min())
print("Max TotalAmount",df['TotalAmount'].max())
print("Mean TotalAmount:",df['TotalAmount'].mean())
print("Mode:",df['TotalAmount'].median())
print("Mode:",df['TotalAmount'].mode())
print("standard deviation:",df['TotalAmount'].std())
```

TIME SERIES ANALYSIS

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['TotalAmount'],marker='o')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('TotalAmount')
plt.show()
```

Customer and product analysis

```python
gender_count=df['Gender'].value_counts()
print("Gender distribution:")
print("Gender count")
```

```python
gender_spend=df.groupby('Gender')['TotalAmount'].sum()
print("Total Spending Gender:")
print(gender_spend)
```

```python
category_quantity=df.groupby('Product Category')['TotalAmount'].sum()
print('Total sales by product category')
print(category_quantity)
```

```python
top_customers=df.groupby('CustomerID')['TotalAmount'].sum().sort_values(ascending=False).head(5)
print('top 5 customers by spending')
print(top_customers)
```

```python
df['Month']=df['Date'].dt.to_period('M')
Monthly_sales = df.groupby('Month')['TotalAmount'].sum()
print("Monthly sales Trend:")
print(Monthly_sales)
```

```python
gender_spend.plot(kind='bar', title='Total Spending by Gender', color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```

```python
category_sales.plot(kind='bar', title='Total Sales by Product Category', color='orange')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.show()
```

```python
Monthly_sales.plot(kind='line', title='Monthly Sales Trend', marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

Visualization: Present insights through bar charts, line plots, and heatmaps.

```python
import seaborn as sns
```

```python
#Barchart
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Product Category',y='TotalAmount',estimator='sum',ci=None,palette='viridis')
plt.title('Total sales by product')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

Line Plot

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x='Date',y='TotalAmount',marker='o',color='teal')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('TotalSales')
plt.grid(True)
plt.show()
```

```python
#BarChart
plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Gender',y='TotalAmount',estimator='sum',ci=None,palette='magma')
plt.title('Total sales by gender')
plt.xlabel('Gender')
plt.ylabel('Total sales')
plt.show()
```

```python
#HEATMAP
plt.figure(figsize=(6,4))
sns.heatmap(df[['Age','Quantity','Price','TotalAmount']].corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heat Map')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.pie(df['Gender'].value_counts(),labels=df['Gender'].value_counts().index,autopct='%1.2f%%',colors=['grey','green'])
plt.title('Gender Distribution',fontsize=17)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Age'],bins=20,kde=True)
```

```python
correlation = df[df.dtypes[df.dtypes == 'int64'].index].corr()
sns.set(font_scale=1.5)
plt.figure(figsize=(12,6))
ax = sns.heatmap(correlation,annot=True,annot_kws={'size': 14},fmt='.1f')
plt.title('Correlation Heatmap',fontsize=17)
plt.show()
```
